```python
data_all=pd.concat([data1_clean,data2_clean])
data_all=data_all.fillna(0) #wNaN 
```

```python
data_all
```

```python
data_all.loc[data_all['total_songs']>6000,'total_songs']=6000
```

```python
# edit city
data_all['city'] = data_all.city.apply(lambda x:"big" if x==1 else "median" if x in[4,5,6,13,14,15,22] else 'small' if x==0 else 'other')
```

```python
data_all['city'].unique()
```

```python
# edit register_via
data_all['registered_via'] = data_all.registered_via.apply(lambda x:'7' if x==7. else '9' if x==9. else '0' if x==0. else 'other')
```

```python
data_all['registered_via'].unique()
```

```python
# edit age
data_all['bd'] = data_all.bd.apply(lambda x: 1 if 3 < x <= 26 else 0 if 26 < x <= 90 else -1)
```

```python
# get dummies
city = pd.get_dummies(data_all['city'],prefix='city')
registered_via = pd.get_dummies(data_all['registered_via'],prefix='registered_via')
bd = pd.get_dummies(data_all['bd'],prefix='bd')

# combine dummies
data_all = pd.concat([data_all,city,registered_via,bd],1)

del data_all['city']
del data_all['registered_via']
del data_all['bd']
```

```python
data_all.columns
```

```python
data_all_userlogs_0=data_all[data_all['have_user_logs']==0]
```

```python
data_all_userlogs_0
```

```python
X_userlogs0=data_all_userlogs_0.drop(columns=['is_churn'])
y_userlogs0=data_all_userlogs_0.is_churn
```

```python
X_userlogs0_train,X_userlogs0_test, y_userlogs0_train, y_userlogs0_test = train_test_split(X_userlogs0, y_userlogs0, test_size=0.25, random_state=42)
```

```python
data_all_userlogs_1=data_all[data_all['have_user_logs']==1]
data_all_userlogs_1
```

```python
X_userlogs1=data_all_userlogs_1.drop(columns=['is_churn'])
y_userlogs1=data_all_userlogs_1.is_churn
```

```python
X_userlogs1_train,X_userlogs1_test, y_userlogs1_train, y_userlogs1_test = train_test_split(X_userlogs1, y_userlogs1, test_size=0.25, random_state=42)
```

```python
X_train=pd.concat([X_userlogs0_train,X_userlogs1_train])
X_test=pd.concat([X_userlogs0_test,X_userlogs1_test])
y_train=pd.concat([y_userlogs0_train,y_userlogs1_train])
y_test=pd.concat([y_userlogs0_test,y_userlogs1_test])
```

```python
from sklearn import linear_model
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.metrics import log_loss

dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
dtc_pred=dtc.predict_proba(X_test)[:,1]

fpr_2,tpr_2,thresholds_2=roc_curve(y_test,dtc_pred)
plt.plot(fpr_2,tpr_2,label='DecisionTreeClassifier(AUC=%0.3f)'%roc_auc_score(y_test,dtc_pred))
plt.plot([0,1],[0,1],'k--')
plt.legend()
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')
```

```python
log_loss(y_test,dtc_pred)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
importances = dtc.feature_importances_
indices = np.argsort(importances)
names = [X_train.columns[i] for i in indices]
plt.figure(figsize=(12,10))
plt.title("Feature Importance")
plt.barh(range(X_train.shape[1]),importances[indices],align='center', alpha=0.5)
plt.yticks(range(X_train.shape[1]), names)
plt.show()
```

```python
X_train_lr = X_train.drop(['city_other','registered_via_other','bd_1'],1)
X_test_lr = X_test.drop(['city_other','registered_via_other','bd_1'],1)
```

```python
from sklearn import linear_model

clf_LR = linear_model.LogisticRegression()
clf_LR.fit(X_train_lr,y_train)
LR_pred=clf_LR.predict_proba(X_test_lr)[:,1]

fpr_1,tpr_1,thresholds_1=roc_curve(y_test,LR_pred)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_1,tpr_1,label='Logistics Regression(AUC=%0.3f)'%roc_auc_score(y_test,LR_pred))

plt.legend()
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')
```

```python
log_loss(y_test,LR_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(OOB=True)
rf.fit(X_train,y_train)
rf_pred=rf.predict_proba(X_test)[:,1]

fpr_1,tpr_1,thresholds_1=roc_curve(y_test,rf_pred)
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr_1,tpr_1,label='Random Forest(AUC=%0.3f)'%roc_auc_score(y_test,rf_pred))
plt.legend()
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')
```
